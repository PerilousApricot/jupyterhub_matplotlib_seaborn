### Note on your /home directory

If you're familiar with the ACCRE environment, you know that you have a persistent home directory at `/home`. The JupyterHub `/home` directory is separate! Because of security restrictions, you can't "see" the other home directory from the other service. To copy files back and forth, open a terminal in JupyterHub and use `scp` to copy files back and forth. A future update to the ACCRE environment will allow users to "see" the Jupyter environment from cluster gateways -- stay tuned for an announcement.

Each user has only 10GB to use on the Jupyter `/home` directory, and is thus unsuitable for very large input files. If you have larger input datasets, they should be stored in our very large HDFS installation. You can find your home directory at `/mnt/hdfs/user/<your username>`

### Note about JupyterHub

The JupyterHub instance you've connected to at `jupyter.accre.vanderbilt.edu` is available to anyone with a valid VUNetID -- not just during this tutorial. If you find it useful, please feel free to use it! It is a shared resource, however, so once you're done with it, please make sure to shutdown your server by clicking "Control Panel" then "Stop My Server".
